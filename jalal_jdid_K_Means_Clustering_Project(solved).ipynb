## K Means Cluster Creation

Now it is time to create the Cluster labels!

** Import KMeans from SciKit Learn.**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import silhouette_score,classification_report
```

```python
X=df.drop('Private',axis=1)
```

#scale features

```python
scale=StandardScaler()
X_scaled=scale.fit_transform(X)
```

### Choose optimal K

```python
sse=[]
for i in range(2,8):
    km=KMeans(n_clusters=i,random_state=42)
    km.fit_predict(X_scaled)
    sse.append(km.inertia_)
    
    print(f'For {i} clusters the silhouette score is {silhouette_score(X_scaled,km.labels_)}')
```

```python
plt.plot(range(2,8),sse)
plt.xlabel('Number of Clusters')
```

### By comparing both SSE error AND silhuette Score BEST K=2

** Create an instance of a K Means model with 2 clusters.**

```python
km=KMeans(n_clusters=2,random_state=42)
labels=km.fit_predict(X_scaled)
```

** What are the cluster center vectors?**

```python
centers=km.cluster_centers_
centers
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters, keep in mind, you usually won't have this luxury in the real world.

** Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
df['cluster']=df['Private'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
df.head()
```

### As we know the Private column here we can make evaluation like in Classification

```python
df['cluster'].value_counts(normalize=True) #imbalanced dataset
```

```python
print(silhouette_score(X_scaled,df.cluster))  #silhouette score for actual labels almost same as the silhouette score for predicted labels
```

```python
print(classification_report(df['cluster'],labels)) #f1 score is 0.52 which is not good for imbalanced dataset
```

# PCA projection

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2,random_state=42)
X_pca=pca.fit_transform(X_scaled)
```

```python
sns.scatterplot(x=X_pca[:,0],y=X_pca[:,1],hue=df['cluster']) #actual labels
```

```python
sns.scatterplot(x=X_pca[:,0],y=X_pca[:,1],hue=labels) #predicted labels from KMeans           
```

# **Naive Bayes Gaussian**

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.data,
                                                    df.target, test_size=0.3,
                                                    random_state=100)
```

```python
nb=GaussianNB()
nb.fit(X_train,y_train)
predictions=nb.predict(X_test)
```

```python
print(predictions)
```

```python
print("Accuraccy :",accuracy_score(y_test,predictions))
```

```python
print("R2 :",nb.score(X_train,y_train))
```

# **Spam** Detection with Multinomial Naive Bayes

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

# Sample dataset
emails = [
    "Win a free lottery now",
    "Hi John, how are you?",
    "Exclusive offer just for you, click now are are",
    "Let's meet for lunch tomorrow"
]
labels = [1, 0, 1, 0]  # 1 = spam, 0 = not spam

# Vectorization
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

# Split
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.3, random_state=42)

# Train model
model = MultinomialNB()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

#
```

```python
vectorizer.get_feature_names_out()
```

# print the token with count in the sparse **matrix**

```python
feature_names = vectorizer.get_feature_names_out()
for i, count in enumerate(X.toarray()[2]):
    if count > 0: # Only print tokens that appear in this email
        print(f"'{feature_names[i]}': {int(count)}")
```
